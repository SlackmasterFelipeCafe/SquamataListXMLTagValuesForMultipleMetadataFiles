```python
#Set Data Paths - perhaps we'll get a user form to do this some day?
# Set the path between the double quotes using the example below
# The r signifies a string literal. Use for paths.
#************************************************************************************************************************
#**********************************************Set Path Here*************************************************************
xmlPath = r"C:\CurrentWork\DataReleases\NationalWildlifeGroundConductivity\MetaData911"
#************************************************************************************************************************
print ('XML File Path set to: ' + xmlPath)
```

```python
#Now create an array listing all the XML files in the directory set above.
xmlFileList = glob.glob(os.path.join(xmlPath, '*.xml'),  recursive=True)
fileListing = ''
xmlFileListing = []
for i in range(len(xmlFileList)):
    splitxmlFileList = xmlFileList[i].split('\\')
    fileListing = fileListing +  splitxmlFileList[len(splitxmlFileList) - 1] + '\n'
    xmlFileListing.append(splitxmlFileList[len(splitxmlFileList) - 1])
#print (fileListing)
xmlFileListing
```

```python
#Load EDI File, Read EDI File to text variable , Close EDI file.
#Locate and print bounding coordinates line by line from text variable input.
    
for i in range(len(xmlFileListing)):
    xmlFile = open(xmlPath + '\\' + xmlFileListing[i], 'r')
    xmlContent = xmlFile.read()
    xmlFile.close()
    #print(xmlContent)
#Now assign values to the bounding coordinates by searching the text variable line by line
# and splitting the approprate string to get corresponding value.
    list_ = xmlContent.splitlines()
    list_length = len (list_)
    print(xmlFileListing[i])
    for X in list_:
        if "<title>" in X:
            titleArray = X.split('>')
            titleValArray = titleArray[1].split('<')
            titleVal = titleValArray[0]
            print ('TITLE: ' + titleVal)
        elif "<westbc>" in X:
            westbcArray = X.split('>')
            westbcValArray = westbcArray[1].split('<')
            westbcVal = westbcValArray[0]
            print ('<westbc>' + westbcVal)
        elif "<eastbc>" in X:
            eastbcArray = X.split('>')
            eastbcValArray = eastbcArray[1].split('<')
            eastbcVal = eastbcValArray[0]
            print ('<eastbc>' + eastbcVal)
        elif "<northbc>" in X:
            northbcArray = X.split('>')
            northbcValArray = northbcArray[1].split('<')
            northbcVal = northbcValArray[0]
            print ('<northbc>' + northbcVal)
        elif "<southbc>" in X:
            southbcArray = X.split('>')
            southbcValArray = southbcArray[1].split('<')
            southbcVal = southbcValArray[0]
            print ('<southbc>' + southbcVal)

  
```

```python
#Load EDI File, Read EDI File to text variable , Close EDI file.
#Locate and print bounding coordinates line by line from text variable input.
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'
    
for i in range(len(xmlFileListing)):
    xmlFile = open(xmlPath + '\\' + xmlFileListing[i], 'r')
    xmlContent = xmlFile.read()
    xmlFile.close()
    #print(xmlContent)
#Now assign values to the bounding coordinates by searching the text variable line by line
# and splitting the approprate string to get corresponding value.
    list_ = xmlContent.splitlines()
    list_length = len (list_)
    print('\n')
    print(color.BOLD + 'Metadata File:\n' + color.END + color.GREEN + xmlFileListing[i] + color.END)
    print(color.BOLD + 'CSV Files referenced:' + color.END)
    if i == 0:
        print(color.RED + csvFileListing[0]+ color.END)
        print(color.RED + csvFileListing[1]+ color.END)
    if i > 0:
        print(color.RED + csvFileListing[(2*i)]+ color.END)
        print(color.RED + csvFileListing[(2*i)+1]+ color.END)
    for X in list_:
        if "<title>" in X:
            titleArray = X.split('>')
            titleValArray = titleArray[1].split('<')
            titleVal = titleValArray[0]
            print (color.BOLD + 'TITLE:\n' + color.END + color.UNDERLINE + titleVal + color.END + color.BOLD + '\nE&A:' + color.END)
        elif "<eadetcit>" in X:
            eadetcitArray = X.split('>')
            eadetcitValArray = eadetcitArray[1].split('<')
            eadetcitVal = eadetcitValArray[0]
            print (color.RED + eadetcitVal + color.END)
        
```
